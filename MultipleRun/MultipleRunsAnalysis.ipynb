### 10000 toy dataset

To get more accurate result the same analysis is made with far more toy. This procedure is computationally demanding, so the following result is obtained with parallel computing

```python
all_freq=np.linspace(10353286068+16*651, 10353286068+31*16*651,30)
all_freq=all_freq[:7]
mu_init=[15,20,15,20,15,30,15]
muCI_results=stat.stat_test_parallel(runsData,all_freq,mu_init,len(mu_init))
result_1 = muCI_results[0]
for i in range(1,7):
    result_1 = np.concatenate((result_1, muCI_results[i]))

result_1 = result_1.reshape(7,3)
```
